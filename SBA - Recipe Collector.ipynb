```python
def find_ingredient(searchvals, ing_list):
    return all([any(val in string for string in ing_list) for val in searchvals])
```

```python
def find_ing(searchvals, ing_list):
    return all(item in ing_list for item in searchvals) | all(item + 's' in ing_list for item in searchvals) | all(item[:-1] in ing_list for item in searchvals)
```

```python
def clean_ing(ings):
    clean = [ing.strip() for ing in ings]
    clean = [ing.strip(',') for ing in clean]
    clean = [ing.strip('*') for ing in clean]
    return clean
```

## Sally's Baking Addiction

```python
##Sallys baking addiction
df = pd.DataFrame(columns = ['Blog', 'Recipe', 'Link', 'Time', 'Ingredients'])
sba_links = []
for i in range(1,38):
    url = f'https://sallysbakingaddiction.com/recipes/?fwp_paged={i}'
    req = Request(url , headers={'User-Agent': 'Mozilla/5.0'})
    sleep(randint(1,2))
    webpage = urlopen(req).read()
    page_soup = soup(webpage, "html.parser")
    content = page_soup.find('div', class_='c-content-panel')
    links = content.find_all('a')
    for link in links:
        sba_links.append(link['href'])
```

```python
for link in sba_links:
    print(link)
    try:
        req = Request(link , headers={'User-Agent': 'Mozilla/5.0'})
        sleep(randint(1,2))
        webpage = urlopen(req).read()
        page_soup = soup(webpage, "html.parser")
        title = page_soup.find('h1').text
        time = page_soup.find('span', class_ = 'tasty-recipes-total-time').text
        ingredients_content = page_soup.find('div', class_ = 'tasty-recipes-ingredients')
        ingredients = ingredients_content.find_all('strong')
        ingredients = [ingredient.text.lower().strip('*,') for ingredient in ingredients]
        df.loc[len(df.index)] = ['Sally\'s baking addiction', title, link, time, ingredients] 
        #df.append({'Blog': 'Sally\'s baking addiction', 'Recipe': title, 'Link': link, 'Time': time, 'Ingredients': ingredients}, ignore_index=True)
    except:
        continue
```

```python
df.Ingredients = df.Ingredients.apply(clean_ing)
df.Ingredients
```

```python
df.to_csv(r'C:\Users\tanne\OneDrive\Documents\SBA_recipes.csv', index = False, header=True)
```

## Ambitious Kitchen

```python
##Ambitious kitchen
df = pd.DataFrame(columns = ['Blog', 'Recipe', 'Link', 'Time', 'Ingredients'])
ak_links = []
for i in range(1,38):
    url = f'https://www.ambitiouskitchen.com/recipe-index/?sf_paged={i}'
    req = Request(url , headers={'User-Agent': 'Mozilla/5.0'})
    sleep(randint(1,2))
    webpage = urlopen(req).read()
    page_soup = soup(webpage, "html.parser")
    ak_content = page_soup.find_all('h2')
    for link in ak_content:
        ak_links.append(link.find('a')['href'])
```
